```python
import sys
import os
import cv2
from pathlib import Path

from ultralytics import YOLO
```

```python
# Find pyproject.toml in parent directories and add its directory to sys.path
current_path = Path.cwd()
for parent in [current_path] + list(current_path.parents):
    pyproject_file = parent / "pyproject.toml"
    if pyproject_file.exists():
        sys.path.insert(0, str(parent))
        break
```

```python
# Create output directory and save result images
output_dir = current_path.parent / "output"
output_dir.mkdir(exist_ok=True)
```

```python
# Load a pre-trained YOLO model (e.g., YOLOv8s for small, yolov8n for nano)
model = YOLO("yolo12l.pt") 
```

```python
img_path: Path = Path("../images/ampan.jpg")
```

```python
# Perform object detection on an image
results = model.predict(source=img_path)
```

```python
# Process and display results (e.g., show bounding boxes)
for r in results:
    r.show() 
```

```python
# Get original image filename without extension and replace spaces with underscores

orig_img_path = Path(r.path)
folder_name = orig_img_path.stem.replace(" ", "_")
result_folder = output_dir / folder_name
result_folder.mkdir(exist_ok=True)

# Save result image(s) to the new folder
for i, res in enumerate(results):
    save_path = output_dir / f"result_{orig_img_path.stem.replace(" ", "_")}.jpg"
    res.save(filename=str(save_path))
```
