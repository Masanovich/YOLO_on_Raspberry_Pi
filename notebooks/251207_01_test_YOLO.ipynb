```python
import sys
import os
import cv2
from pathlib import Path

from ultralytics import YOLO
```

```python
# Find pyproject.toml in parent directories and add its directory to sys.path
path_curr = Path.cwd()
for parent in [path_curr] + list(path_curr.parents):
    path_pyproject = parent / "pyproject.toml"
    if path_pyproject.exists():
        sys.path.insert(0, str(parent))
        break
```

```python
# Create output directory and save result images
dir_output = path_curr.parent / "output"
dir_output.mkdir(exist_ok=True)
```

```python
# Load a pre-trained YOLO model (e.g., YOLOv8s for small, yolov8n for nano)
model = YOLO("yolo12l.pt") 
```

```python
img_path: Path = Path("../images/syamu.jpg")
```

```python
# Perform object detection on an image
results = model.predict(source=img_path)
```

```python
# Process and display results (e.g., show bounding boxes)
for r in results:
    r.show() 
```

```python
# Get original image filename without extension and replace spaces with underscores

path_image_origal = Path(r.path)
dir_save = dir_output / path_image_origal.stem.replace(" ", "_")
dir_save.mkdir(exist_ok=True)

# Save result image(s) to the new folder
for i, res in enumerate(results):
    save_path = dir_save / f"result_{path_image_origal.stem.replace(" ", "_")}.jpg"
    res.save(filename=str(save_path))
```
